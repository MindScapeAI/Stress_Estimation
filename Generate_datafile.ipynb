```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ETRI_path = '/raid/mnt/Yeoneui/ETRI'
```

```python
dir2018_path = "/raid/mnt/gahee/ETRI_dataset/dataset_2018"
dir2019_path = "/raid/mnt/gahee/ETRI_dataset/dataset_2019"
dir2020_path1 = "/raid/mnt/gahee/ETRI_dataset/user01-06"
dir2020_path2 = "/raid/mnt/gahee/ETRI_dataset/user07-10"
dir2020_path3 = "/raid/mnt/gahee/ETRI_dataset/user11-12"
dir2020_path4 = "/raid/mnt/gahee/ETRI_dataset/user21-25"
dir2020_path5 = "/raid/mnt/gahee/ETRI_dataset/user26-30"
```

```python
print(users2018)
```

```python
#2018

for user in users2018:
    if not os.path.exists(os.path.join(ETRI_path, '2018_user',user)):
        os.makedirs(os.path.join(ETRI_path, '2018_user',user))
    days = os.listdir(os.path.join(dir2018_path,user))
    for day in days:
        if os.path.exists(f'{dir2018_path}/{user}/{day}/{day}_label.csv'):
            df = pd.read_csv(f'{dir2018_path}/{user}/{day}/{day}_label.csv')
            df_filtered = df[['ts','action','condition','emotionPositive','emotionTension']]
            df_filtered.to_csv(os.path.join(ETRI_path, '2018_user', user, f'{day}_label.csv'))
            
```
